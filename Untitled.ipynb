```python
import pandas as pd
import numpy as np
import torch
```

```python
log = pd.read_csv("driving_log.csv")
```

```python
log
```

```python
# basically swoops all but the last value from the database (which is null anyways so we want to skip it)
waypoints = log.iloc[:, 7:17].to_numpy(dtype=np.unicode_).tolist()
```

```python
for num, values in enumerate(waypoints):
    for ind, wp in enumerate(values):
        answer = [float(v) for v in wp.split(' ')]
        waypoints[num][ind] = answer
```

```python
complete = np.array(waypoints)
print(complete[0])
```
